```python
import re
```

```python
# Transformar os valores das colunas que estão erradas em vazias para deletalas

df['ANO_EXECUCAO'] = df['ANO_EXECUCAO'].apply(lambda x: re.findall(r'\d{4}', str(x))[0] if re.findall(r'\d{4}', str(x)) else 'None')

print(df)
```

```python
# Removendo as linhas contendo dados atribuidos com None

filtered_dados = df[df['ANO_EXECUCAO'] != 'None']

print(filtered_dados)
```

```python
# Verificando se os erros foram removidos

filtered_dados['ANO_EXECUCAO'].unique()
```

```python
filtered_dados['AREA_TEMATICA'].unique()
```

```python
# Filtrando os dados para remover linhas vazias juntamente com as linhas que estão erradas

filtered_dados.dropna(subset=['ANO_EXECUCAO'], axis=0)
```

```python
# Atribuindo os dados a uma nova variável

dados = filtered_dados.dropna(subset=['ANO_EXECUCAO'], axis=0)
```

```python
# Removendo a coluna vazia

dados.drop('Unnamed: 7', axis= 1, inplace = True)

dados
```

```python
# Salvando o CSV do dataset

dados.to_csv('Aprende_IMAP_Dataset')
```
